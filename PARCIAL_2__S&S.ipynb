```python
%%writefile 1_üìà_EJERCICIO_1.py

import streamlit as st
import time
import numpy as np
import matplotlib.pyplot as plt

st.set_page_config(page_title="Plotting Demo", page_icon="üìà")

# Par√°metros iniciales
Fo = 60  # Frecuencia de la se√±al
T = 1 / Fo  # Periodo
Fs = 100 * Fo  # Frecuencia de muestreo definida
tv = np.arange(-T / 2, T / 2, 1 / Fs)  # Vector de tiempo generado
Nm = len(tv)  # N√∫mero de muestras requeridas
A = 1  # Amplitud de la se√±al
x1 = (A * np.cos(2 * np.pi * Fo * tv)) ** 2  # Se√±al cuadr√°tica

# N√∫mero de arm√≥nicos
N = 50  # Cantidad de t√©rminos en la serie de Fourier
wo = 2 * np.pi / T  # Frecuencia fundamental

# Definir bases de Fourier
phin = np.zeros((Nm, 2 * N + 1), dtype=np.complex_)
for n in range(-N, N + 1, 1):
    phin[:, n + N] = np.exp(1j * n * wo * tv)  # Base de Fourier

# Calcular espectro
nv = np.linspace(-N, N, 2 * N + 1)  # Vector num√©rico de arm√≥nicos
Cn = np.zeros(2 * N + 1, dtype=np.complex_)
Cn[N] = (A ** 2) / 2  # Nivel DC
Cn[N + 2] = (A ** 2) / 4  # Primer arm√≥nico positivo
Cn[N - 2] = (A ** 2) / 4  # Primer arm√≥nico negativo

# Interfaz Streamlit
st.markdown("<h1 style='text-align: center;'>An√°lisis de Fourier de una Se√±al</h1>", unsafe_allow_html=True)

# Mostrar se√±al original
st.subheader("Se√±al Original")
st.write( "Esta gr√°fica muestra la se√±al original en el dominio del tiempo, una se√±al coseno elevada al cuadrado, que resulta en una forma de onda peri√≥dica con frecuencia 60 Hz."
)
fig, ax = plt.subplots()
ax.plot(tv, x1, 'r', linewidth=4)
ax.set_xlabel("t [s]")
ax.set_ylabel("x(t)")
ax.grid()
st.pyplot(fig)

# Graficar espectro
st.subheader("Espectro de Fourier")
st.write("El espectro de Fourier descompone la se√±al en sus componentes arm√≥nicas. Aqu√≠ se representan la parte real, imaginaria, magnitud y fase de los coeficientes de Fourier.")

fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0, 0].stem(nv, np.real(Cn), 'r')
axs[0, 0].set_title("Parte Real")
axs[0, 0].set_xlabel("n")
axs[0, 0].set_ylabel("Re{C_n}")
axs[0, 0].grid()

axs[0, 1].stem(nv, np.imag(Cn), 'r')
axs[0, 1].set_title("Parte Imaginaria")
axs[0, 1].set_xlabel("n")
axs[0, 1].set_ylabel("Im{C_n}")
axs[0, 1].grid()

axs[1, 0].stem(nv, abs(Cn), 'r')
axs[1, 0].set_title("Magnitud")
axs[1, 0].set_xlabel("n")
axs[1, 0].set_ylabel("|C_n|")
axs[1, 0].grid()

axs[1, 1].stem(nv, np.angle(Cn), 'r')
axs[1, 1].set_title("Fase")
axs[1, 1].set_xlabel("n")
axs[1, 1].set_ylabel("‚à†C_n")
axs[1, 1].grid()

fig.tight_layout()
st.pyplot(fig)

# Graficar espectro en dB y fase
st.subheader("Magnitud y Fase en dB")
st.write("Aqu√≠ se representa la magnitud en escala logar√≠tmica (dB) y la fase en grados, mostrando la relevancia de cada arm√≥nico en la reconstrucci√≥n de la se√±al.")
fig, axs = plt.subplots(2, 1, figsize=(8, 8))
axs[0].stem(nv, 20 * np.log10(abs(Cn)), 'r')
axs[0].set_title("Magnitud en dB")
axs[0].set_xlabel("n")
axs[0].set_ylabel("|C_n| [dB]")
axs[0].grid()

axs[1].stem(nv, np.angle(Cn, deg=True), 'r')
axs[1].set_title("Fase en Grados")
axs[1].set_xlabel("n")
axs[1].set_ylabel("Fase [¬∞]")
axs[1].grid()

fig.tight_layout()
st.pyplot(fig)

# Interactivo: Se√±al reconstruida
st.subheader("Reconstrucci√≥n de la Se√±al")
st.write("Se reconstruye la se√±al usando una cantidad variable de arm√≥nicos. A medida que aumentamos los arm√≥nicos, la reconstrucci√≥n se acerca m√°s a la se√±al original.")
Na = st.slider("N√∫mero de arm√≥nicos", 0, N, 1)
ind = np.arange(N - Na, N + Na + 1)
ind = np.clip(ind, 0, len(Cn) - 1)
Px = 0.375  # Estimar seg√∫n se√±al estudiada
er = 1 - np.sum(abs(Cn[ind]) ** 2) / Px
xe = phin[:, ind] @ Cn[ind].ravel()

fig, ax = plt.subplots()
ax.plot(tv, xe, 'b', marker='o', markersize=5, label='$x_e(t)$')
ax.plot(tv, x1, 'r', label='x(t)')
ax.set_title(f'$E_r$={100 * er:.2f}%')
ax.set_xlabel("t [s]")
ax.set_ylabel("x(t)")
ax.grid()
ax.legend()
st.pyplot(fig)

```

```python
%%writefile 2_üéµüìä_EJERCICIO_2.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import yt_dlp
import soundfile as sf
from io import BytesIO
import tempfile
import os
from scipy.io.wavfile import write
import io

# Configuraci√≥n inicial
st.title("Modulaci√≥n AM")

# Par√°metros
Fc = 100  # Frecuencia de la portadora (Hz)
Fm = 10   # Frecuencia del mensaje (Hz)
Am = 1    # Amplitud del mensaje

# Definici√≥n del tiempo
ti = 0
tf = 5 / Fm  # N√∫mero de per√≠odos del mensaje a graficar
Fs = 5 * Fc  # Frecuencia de muestreo
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo

# Generaci√≥n de se√±ales
m = Am * np.cos(2 * np.pi * Fm * t)  # Se√±al del mensaje
Im = 0.8  # √çndice de modulaci√≥n
Ac = max(abs(m)) / Im  # Amplitud de la portadora seg√∫n el √≠ndice de modulaci√≥n
c = Ac * np.sin(2 * np.pi * Fc * t)  # Se√±al portadora

y = (1 + m / Ac) * c  # Se√±al modulada en AM

# Gr√°fico 1: Se√±al de la portadora y mensaje
st.subheader("Se√±al de la portadora y mensaje")
fig1, ax1 = plt.subplots()
ax1.plot(t, c, label='Portadora')
ax1.plot(t, m, label='Mensaje')
ax1.set_xlabel("Tiempo (s)")
ax1.set_ylabel("Amplitud")
ax1.set_title("Portadora y mensaje en el tiempo")
ax1.legend()
st.pyplot(fig1)

# Explicaci√≥n del gr√°fico 1
st.write("Este gr√°fico muestra la se√±al de la portadora (onda senoidal de alta frecuencia) y la se√±al del mensaje (onda cosenoidal de menor frecuencia). La portadora sirve como la base sobre la cual se modula la informaci√≥n del mensaje.")

# Gr√°fico 2: Se√±al modulada en AM
st.subheader("Se√±al modulada en AM")
fig2, ax2 = plt.subplots()
ax2.plot(t, y, label='Se√±al AM')
ax2.set_xlabel("Tiempo (s)")
ax2.set_ylabel("Amplitud")
ax2.set_title("Se√±al modulada en amplitud")
ax2.legend()
st.pyplot(fig2)

# Explicaci√≥n del gr√°fico 2
st.write("En este gr√°fico se muestra la se√±al resultante despu√©s de la modulaci√≥n en amplitud (AM). La envolvente de la se√±al modulada sigue la forma de la se√±al del mensaje, lo que permite recuperar la informaci√≥n original en el receptor.")

# Transformada de Fourier
Xfc = np.fft.fft(c)
Xfm = np.fft.fft(m)
Xfy = np.fft.fft(y)

vfre = np.fft.fftfreq(len(c), 1 / Fs)

# Gr√°fico 3: Espectro de frecuencia
st.subheader("Espectro de frecuencia de las se√±ales")
fig3, ax3 = plt.subplots()
ax3.plot(vfre, abs(Xfy), label='Modulado')
ax3.plot(vfre, abs(Xfm), label='Mensaje')
ax3.plot(vfre, abs(Xfc), label='Portadora')
ax3.set_xlabel("Frecuencia (Hz)")
ax3.set_ylabel("|X(f)|")
ax3.set_title("An√°lisis espectral de las se√±ales")
ax3.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])
ax3.legend()
st.pyplot(fig3)

# Explicaci√≥n del gr√°fico 3
st.write("Este gr√°fico muestra el espectro de frecuencia de las se√±ales. Se puede observar la se√±al del mensaje en bajas frecuencias, la portadora en su frecuencia correspondiente y la se√±al modulada con sus componentes espectrales en torno a la portadora.")

#Canci√≥n

# Configuraci√≥n de la app
st.title("Descarga y reproducci√≥n de audio desde YouTube")

# Informaci√≥n de la canci√≥n
nombre_cancion = "BAD BUNNY - VOY A LLeVARTE PA PR (Visualizer) | DeB√ç TiRAR M√°S FOToS"
st.write(f"**Canci√≥n:** {nombre_cancion}")

# Descarga del audio desde YouTube
def descargar_audio(url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'extractaudio': True,
        'audioformat': 'mp3',
        'outtmpl': 'tema.mp3'
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

def convertir_a_wav():
    os.system("ffmpeg -y -i tema.mp3 output.wav")

# URL del video de YouTube
link = "https://www.youtube.com/watch?v=X2MGCIDOMZ4"
st.write("Descargando audio...")
descargar_audio(link)
st.write("Convirtiendo a WAV...")
convertir_a_wav()

# Cargar el audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

# Mostrar la frecuencia de muestreo
st.write(f"**Frecuencia de muestreo:** {fs} Hz")

# Selecci√≥n del fragmento de audio (segundos 20 a 25)
inicio, fin = int(fs * 20), int(fs * 25)
segmento_audio = x[inicio:fin]

# Guardar el segmento en un archivo temporal
with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as temp_wav:
    sf.write(temp_wav.name, segmento_audio, fs)
    temp_wav_path = temp_wav.name

# Reproducir el audio en Streamlit
st.audio(temp_wav_path, format="audio/wav")

st.write("Este audio corresponde al segmento entre los segundos 20 y 25 de la canci√≥n.")

# Configuraci√≥n de la app
st.title("Procesamiento de Audio y Modulaci√≥n AM")

# Cargar el audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

# Seleccionar el canal izquierdo y fragmento de audio (segundos 20 a 25)
ti, tf = 20, 25
xpro = x.copy()
xs = xpro[int(ti * fs):int(tf * fs), 0]

# Vector de tiempo
tt = np.arange(ti, tf, 1 / fs)

# Gr√°fico 1: Se√±al original (canal izquierdo)
st.subheader("Se√±al Original del Canal Izquierdo")
fig1, ax1 = plt.subplots()
ax1.plot(tt, xs, label="Canal izquierdo")
ax1.set_xlabel("Tiempo (s)")
ax1.set_ylabel("Amplitud")
ax1.set_title("Se√±al de audio entre 20s y 25s")
ax1.legend()
st.pyplot(fig1)

# Explicaci√≥n del gr√°fico 1
st.write("Este gr√°fico muestra la se√±al original del canal izquierdo en el intervalo de tiempo entre 20 y 25 segundos.")

# Entrada del usuario para el √≠ndice de modulaci√≥n
Im = st.number_input("Ingrese el √≠ndice de modulaci√≥n:", min_value=0.1, max_value=2.0, value=0.8, step=0.1)

# Par√°metros de modulaci√≥n AM
Fc = 1000  # Frecuencia de la portadora (Hz)
Fs = 48000  # Frecuencia de muestreo
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo

# Generar la portadora
Ac = max(abs(xs)) / Im  # Amplitud de la portadora seg√∫n √≠ndice de modulaci√≥n
c = Ac * np.sin(2 * np.pi * Fc * t)  # Se√±al portadora

# Gr√°fico 2: Se√±al portadora y mensaje
st.subheader("Se√±al Portadora y Mensaje")
fig2, ax2 = plt.subplots()
ax2.plot(t, c, label='Portadora')
ax2.plot(t, xs, label='Mensaje')
ax2.set_xlabel("Tiempo (s)")
ax2.set_ylabel("Amplitud")
ax2.set_title("Comparaci√≥n entre portadora y mensaje")
ax2.legend()
st.pyplot(fig2)

# Explicaci√≥n del gr√°fico 2
st.write("En este gr√°fico se observa la se√±al portadora (onda senoidal de alta frecuencia) y la se√±al del mensaje original (audio extra√≠do). La portadora se usa para modular la se√±al de audio.")

# Se√±al modulada en AM
y = (1 + xs / Ac) * c

# Gr√°fico 3: Se√±al modulada AM
st.subheader("Se√±al Modulada en AM")
fig3, ax3 = plt.subplots()
ax3.plot(t, y, label='Se√±al Modulada AM')
ax3.set_xlabel("Tiempo (s)")
ax3.set_ylabel("Amplitud")
ax3.set_title("Modulaci√≥n AM de la Se√±al de Audio")
ax3.legend()
st.pyplot(fig3)

st.write("Este gr√°fico muestra la se√±al modulada en amplitud (AM). La forma de la se√±al refleja c√≥mo la envolvente sigue la se√±al del mensaje, lo que permite transmitir la informaci√≥n del audio.")

# Transformada de Fourier
Xfc = np.fft.rfft(c)
Xfm = np.fft.rfft(xs)
Xfy = np.fft.rfft(y)
vfre = np.fft.rfftfreq(len(c), 1 / Fs)

# Gr√°fico 4: Espectro de la portadora
st.subheader("Espectro de Frecuencia de la Portadora")
fig4, ax4 = plt.subplots()
ax4.plot(vfre, abs(Xfc), label='Portadora')
ax4.set_xlabel("Frecuencia (Hz)")
ax4.set_ylabel("|X(f)|")
ax4.set_title("Espectro de Frecuencia de la Se√±al Portadora")
ax4.legend()
st.pyplot(fig4)
st.write("La se√±al portadora es una onda sinusoidal pura de frecuencia Fc = 1000 Hz. Su espectro de frecuencia muestra un √∫nico pico en esta frecuencia, ya que una onda sinusoidal tiene toda su energ√≠a concentrada en una sola frecuencia")

# Gr√°fico 5: Espectro de la Se√±al Modulada AM
st.subheader("Espectro de Frecuencia de la Se√±al Modulada")
fig5, (ax5a, ax5b) = plt.subplots(2, 1, figsize=(8, 8), constrained_layout=True)  # Aumentando el espacio
ax5a.plot(vfre, abs(Xfy), label='Modulado', color='r')
ax5a.set_xlabel("Frecuencia (Hz)")
ax5a.set_ylabel("|X(f)|")
ax5a.set_title("Espectro de la Se√±al Modulada AM")
ax5a.legend()
ax5a.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])

ax5b.plot(vfre, abs(Xfm), label='Mensaje')
ax5b.set_xlabel("Frecuencia (Hz)")
ax5b.set_ylabel("|X(f)|")
ax5b.set_title("Espectro de la Se√±al del Mensaje")
ax5b.legend()
ax5b.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])

st.pyplot(fig5)

st.write("El primer gr√°fico del espectro muestra la se√±al modulada en AM, donde se pueden ver las bandas laterales. El segundo gr√°fico muestra el espectro del mensaje original antes de la modulaci√≥n.")

# Funci√≥n para convertir el audio en un buffer
def convert_audio(data, rate):
    buffer = io.BytesIO()
    write(buffer, rate, data.astype(np.float32))
    buffer.seek(0)  # Reiniciar el puntero para asegurar la reproducci√≥n
    return buffer

# Convertir los audios antes de mostrarlos para reducir la demora
buffer_c = convert_audio(c, fs)
buffer_xs = convert_audio(xs, fs)
buffer_y = convert_audio(y, fs)

# Crear pesta√±as para seleccionar el audio
tab1, tab2, tab3 = st.tabs(["Portadora", "Mensaje", "Modulada"])

with tab1:
    st.subheader("Audio de la portadora")
    st.audio(buffer_c, format='audio/wav')

with tab2:
    st.subheader("Audio del mensaje original")
    st.audio(buffer_xs, format='audio/wav')

with tab3:
    st.subheader("Audio de la se√±al modulada")
    st.audio(buffer_y, format='audio/wav')

#PARTE ESCRITA

# URL de la carpeta de im√°genes en Google Drive
drive_link = "https://drive.google.com/drive/u/0/folders/1OK3c9fOjyQxqKJdMXZKkVEqcJvQaQzVl"

# CSS para hacer el bot√≥n flotante
st.markdown(f"""
    <style>
    .floating-button {{
        position: fixed;
        bottom: 20px;
        right: 20px;
        background-color: #FF4B4B;
        color: white;
        padding: 10px 20px;
        border-radius: 25px;
        font-size: 18px;
        font-weight: bold;
        text-align: center;
        box-shadow: 2px 2px 10px rgba(0,0,0,0.3);
        cursor: pointer;
        transition: background-color 0.3s ease;
    }}
    .floating-button:hover {{
        background-color: #FF1F1F;
    }}
    </style>
    <a href="{drive_link}" target="_blank">
        <div class="floating-button">üì∏ Ver Parte escrita</div>
    </a>
""", unsafe_allow_html=True)
```

```python
%%writefile 3_üìà_EJERCICIO_3.py

import streamlit as st
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt


st.title("An√°lisis de la Distorsi√≥n Arm√≥nica Total (THD) y su Impacto en el Factor de Potencia")

st.markdown("La primera partede del ejercicio tres se en cuentra en el siguiente pdf: [Solucion teorica](https://drive.google.com/file/d/1LfKPqXS0-0PR8wcRxVwIi02X2JBbDUN1/view?usp=sharing)")


st.subheader("Carga Netamente Resistiva")

st.markdown("Para este caso, el circuito responde a:")


st.latex(r"V_R(s) = R I_{in}(s)")

st.latex(r"\frac{V_R(s)}{V_{in}(s)} = 1")


# Entrada de usuario para el valor de R
R_v = st.number_input("Ingrese el valor de R:", min_value=0.1, value=10.0)

# Par√°metros de la simulaci√≥n
Fo = 60  # Frecuencia de alimentaci√≥n
Fs = 30 * Fo  # Frecuencia de muestreo
To = 1 / Fo  # Per√≠odo fundamental
Ts = 1 / Fs  # Per√≠odo de muestreo
t = np.arange(0, 5 * To, Ts)  # Simulaci√≥n de 5 per√≠odos

A = 120  # Amplitud de la entrada

# Funci√≥n de entrada: onda senoidal rectificada completa
in_o = A * np.sin(2 * np.pi * Fo * t)
rec_c = sig.square(2 * np.pi * Fo * t)  # Tren de pulsos
in_ = in_o * rec_c  # Entrada rectificada completa

# Definir la funci√≥n de transferencia
num = np.array([1])  # Coeficientes del numerador
den = np.array([1])  # Coeficientes del denominador
G_n = sig.TransferFunction(num, den)  # Sistema de primer orden

# Calcular la respuesta del sistema usando lsim
_, out, _ = sig.lsim(G_n, U=in_, T=t)

# --- Gr√°fica 1: Se√±al de alimentaci√≥n ---

st.subheader("Se√±al de Alimentaci√≥n")

st.markdown("Esta gr√°fica muestra la se√±al de alimentaci√≥n antes de ser rectificada. Es una onda senoidal con una frecuencia de 60 Hz y una amplitud de 120V.")

fig, ax = plt.subplots()
ax.plot(t, in_o, label='Alimentaci√≥n')
ax.legend()
st.pyplot(fig)

# --- Gr√°fica 2: Entrada rectificada ---

st.subheader("Se√±al de Entrada Rectificada")

st.markdown("Aqu√≠ se observa la se√±al de entrada despu√©s de ser rectificada. Se puede notar que la parte negativa de la onda senoidal ha sido eliminada, generando una forma de onda pulsante.")
fig, ax = plt.subplots()
ax.plot(t, in_, label='In(t)')
ax.set_xlabel('t [s]')
ax.set_ylabel('In(t)')
ax.legend()
st.pyplot(fig)

# --- Gr√°fica 3: Salida del circuito ---

st.subheader("Se√±al de Salida del Circuito")

st.markdown("La √∫ltima gr√°fica muestra la salida del circuito resistivo. Como el sistema tiene una funci√≥n de transferencia unitaria, la salida es similar a la entrada rectificada.")

fig, ax = plt.subplots()
ax.plot(t, out, c="r", label='Out(t)')
ax.set_xlabel('t [s]')
ax.set_ylabel('Out(t)')
ax.legend()
st.pyplot(fig)

st.markdown("Debido a que la resistencia es una carga no almacenadora de energ√≠a, el voltaje de entrada es igual al voltaje de salida")


st.subheader("Calculo del espectro")

vfre = np.fft.rfftfreq(len(out), 1 / Fs)  # Vector de frecuencia
Xf = np.fft.rfft(out)

# Visualizaci√≥n con Streamlit
st.title("Espectro de Frecuencia")

fig, ax = plt.subplots()
ax.stem(vfre, abs(Xf) / len(out))
ax.set_xlabel("Frecuencia [Hz]")
ax.set_ylabel("|X(f)|")
ax.set_title("Espectro de Frecuencia")
ax.grid()

st.pyplot(fig)

# C√°lculo del THD
def calcular_thd(signal, Fs):
    Xf = np.fft.rfft(signal)  # FFT de la se√±al
    magnitudes = np.abs(Xf)
    freqs = np.fft.rfftfreq(len(signal), 1 / Fs)

    # Identificar la frecuencia fundamental
    fundamental_idx = np.argmax(magnitudes)
    fundamental = magnitudes[fundamental_idx]

    # C√°lculo del THD como la suma de los arm√≥nicos sobre la fundamental
    thd = np.sqrt(np.sum(magnitudes[1:]**2)) / fundamental
    return thd

thd_r = calcular_thd(out, Fs)
st.write(f"**THD del circuito con carga resistiva:** {thd_r * 100:.2f}%")

pf=np.sqrt(1/(1+(thd_r)**2))
st.write(f"El factor de potencia de la se√±al con carga resistiva: {pf:.4f}")

R_v = st.number_input("Ingrese el valor de R:", min_value=0.0, value=1.0, step=0.1)
C_v = st.number_input("Ingrese el valor de C:", min_value=0.0, value=1.0e-6, step=1.0e-6)

# Definir valores de tiempo y se√±al de entrada (deben estar definidos antes de usarlos)
Fs = 1000  # Frecuencia de muestreo
t = np.linspace(0, 0.1, int(0.1 * Fs), endpoint=False)  # Vector de tiempo
in_ = np.abs(np.sin(2 * np.pi * 50 * t))  # Se√±al de entrada rectificada

# Funci√≥n de transferencia del circuito RC
if R_v > 0 and C_v > 0:  # Evita valores inv√°lidos
    num = np.array([1])
    den = np.array([R_v * C_v, 1])
    G_n = sig.TransferFunction(num, den)

    # Evaluar salida ante entrada rectificada mediante respuesta al sistema
    tout, out, _ = sig.lsim(G_n, in_, T=t)

    # Gr√°fico de la se√±al de entrada y salida
    fig, ax = plt.subplots()
    ax.plot(t, in_, label='In(t)')  # Entrada rectificada
    ax.plot(t, out, label='Out(t)')  # Salida del circuito
    ax.set_xlabel('t [s]')
    ax.set_ylabel('out(t)')
    ax.legend()
    st.pyplot(fig)

    # Calcular la FFT
    vfre = np.fft.rfftfreq(len(out), 1 / Fs)  # Vector de frecuencia en Hz
    Xf = np.fft.rfft(out)

    # Crear la gr√°fica del espectro de frecuencia
    fig, ax = plt.subplots()
    ax.stem(vfre, abs(Xf) / len(out), basefmt=" ")  # Evitar l√≠nea base en la gr√°fica
    ax.set_xlabel('Frecuencia [Hz]')
    ax.set_ylabel('|X(f)|')
    ax.set_title('Espectro de Frecuencia')
    ax.grid()
    st.pyplot(fig)

    # Funci√≥n para calcular el THD
    def calcular_thd(signal, Fs):
        Xf = np.fft.rfft(signal)  # FFT de la se√±al
        magnitudes = np.abs(Xf)

        if len(magnitudes) < 2:  # Evita errores si la FFT es demasiado peque√±a
            return 0

        # Identificar la frecuencia fundamental
        fundamental_idx = np.argmax(magnitudes)
        fundamental = magnitudes[fundamental_idx]

        if fundamental == 0:  # Evita divisiones por cero
            return 0

        # C√°lculo del THD como la suma de los arm√≥nicos sobre la fundamental
        thd = np.sqrt(np.sum(magnitudes[1:] ** 2)) / fundamental
        return thd

    # C√°lculo del THD para el circuito RC (con filtro)
    thd_rc = calcular_thd(out, Fs)
    st.write(f"THD del circuito RC: {thd_rc * 100:.2f}%")

    # C√°lculo del factor de potencia
    pf = np.sqrt(1 / (1 + (thd_rc) ** 2))
    st.write(f"El factor de potencia de la se√±al con carga RC: {pf:.4f}")

else:
    st.error("Los valores de R y C deben ser mayores que cero.")
```
